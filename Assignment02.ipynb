```python
#task:02
def unique_users(list):
    unique=[]
    for i in list:
        user_id = i[1]
        if user_id not in unique:
            unique.append(user_id)
    return len(unique)

def highest_transaction(list):
    highest=list[0]
    for i in list:
        if i[2]>highest[2]:
            highest =i
    return highest 

def seperate_lists(list):
    transaction_ids=[]
    user_ids=[]
    for i in list:
        if len(i)>=2:
            transaction_ids.append(i[0])
            user_ids.append(i[1])
        else:
            print(f'inconsistent tuple:{i}')
    return transaction_ids,user_ids
def main():
    list=[(101, 1, 150.50, "2023-01-10 10:00"),
        (102, 2, 200.00, "2023-01-11 11:30"),
        (103, 1, 300.00, "2023-01-12 09:15"),
        (104, 3, 500.50, "2023-01-13 14:00"),
        (105, 4, 100.75, "2023-01-14 16:30"),
        (106, 2, 250.00, "2023-01-15 13:45"),
        (107, 5, 400.00, "2023-01-16 08:20"),
        (108, 6, 600.00, "2023-01-17 11:00")]
    unique_users_count=unique_users(list)
    print(f'total no of unique users:{unique_users_count}')

    highest_trans=highest_transaction(list)
    print(f'transaction with the highest amount:{highest_trans}')

    transaction_ids,user_ids=seperate_lists(list)
    print(f'transaction IDs:{transaction_ids}')
    print(f'user IDs:{user_ids}')
main()
    


```

```python
#task:03
def users_visited_both(page_a,page_b):
    return page_a.intersection(page_b)

def visited_one_page(page_a,page_c):
    return page_a.symmetric_difference(page_c)

def update_page_a(page_a,new_users):
    page_a.update(new_users)

def main():
    page_a={1,2,3,4,5,6}
    page_b={4,5,6,7,8}
    page_c={6,7,9,10}

    both=users_visited_both(page_a,page_b)
    print(f'users who visited both page A and B:{both}')

    a=visited_one_page(page_a,page_c)
    print(f'users who visited either page A or C but not both:{a}')

    new_users={7,8,9}
    update_page_a(page_a,new_users)
    print(f'updated page A user IDs:{page_a}')
main()
```

```python
#task:04
def high_ratings(dict_1):
    high_ratingss = {}
    for user_id, feedback in dict_1.items():
        rating = feedback['rating']
        if rating >= 4:
            high_ratingss[user_id] = rating
    return high_ratingss

def top_5_users_by_rating(high_ratingss):
    sorted_ratings = sorted(high_ratingss.items(), key=lambda x: x[1], reverse=True)
    return sorted_ratings[:5]

def combine_feedback(dicts):
    a = {}
    for feedback_dict in dicts:
        for user_id, feedback in feedback_dict.items():
            rating = feedback['rating']
            comments = feedback['comments']
            if user_id in a:
                a[user_id]['rating'] = max(a[user_id]['rating'], rating)
                a[user_id]['comments'] += " " + comments
            else:
                a[user_id] = {'rating': rating, 'comments': comments}
    return a

def ratings_above_three(dict_1):
    ratings_above_3 = {}
    for user_id, feedback in dict_1.items():
        rating = feedback['rating']
        if rating > 3:
            ratings_above_3[user_id] = rating
    return ratings_above_3

def main():
    dict_1 = {
        1: {'rating': 5, 'comments': 'Excellent service!'},
        2: {'rating': 3, 'comments': 'Average experience.'},
        3: {'rating': 4, 'comments': 'Good, but room for improvement.'},
        4: {'rating': 2, 'comments': 'Not satisfied.'}
    }
    dict_2 = {
        1: {'rating': 4, 'comments': 'Helpful staff.'},
        3: {'rating': 5, 'comments': 'Great improvement!'},
        5: {'rating': 5, 'comments': 'Amazing quality!'},
        6: {'rating': 4, 'comments': 'Nice work.'}
    }
    dict_3 = {
        2: {'rating': 4, 'comments': 'Much better experience.'},
        5: {'rating': 4, 'comments': 'Very satisfied.'}
    }

    a = combine_feedback([dict_1, dict_2, dict_3])
    
    print("Combined Feedback from Users:")
    for user_id, feedback in a.items():
        print(f"User ID {user_id}: Rating {feedback['rating']}, Comments: {feedback['comments']}")
    
    high_ratingss = high_ratings(a)
    print("\nUsers with High Ratings (4 and above):")
    for user_id, rating in high_ratingss.items():
        print(f"User ID {user_id} has a rating of {rating}.")

    top_5 = top_5_users_by_rating(high_ratingss)
    print("\nTop 5 Users by Rating:")
    for user in top_5:
        print(f"User ID {user[0]}: Rating {user[1]}")
    
    ratings_above_3 = ratings_above_three(a)
    print("\nUsers with Ratings Above 3:")
    for user_id, rating in ratings_above_3.items():
        print(f"User ID {user_id}: Rating {rating}")

main()
```
