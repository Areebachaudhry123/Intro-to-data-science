```python
#task 01 
import pandas as pd
df = pd.read_csv('/kaggle/input/employee-performance-and-salary-dataset/Employe_Performance_dataset.csv')
print(df.head(10))
```

```python
#task 02 
print(df.dtypes)
```

```python
#task 03 
rows, columns = df.shape
print(f"Number of rows: {rows}")
print(f"Number of columns: {columns}")
```

```python
#task 04
cols = df[['Name', 'Age', 'Salary']]
print(cols)
```

```python
#task 05 
first_5_rows = df.iloc[:5]
print(first_5_rows)
```

```python
#task 06
sales_department = df.loc[df['Department'] == 'Sales']
print(sales_department)
```

```python
#task 07 
filtered_rows = df.loc[(df['Salary'] > 8000) & (df['Status'] == 'Active')]
print(filtered_rows)
```

```python
#task 08 
filtered_rows = df.loc[(df['Experience'] > 5) & (df['Performance Score'] >= 4)]
print(filtered_rows)
```

```python
#task 09
alternate_rows = df.iloc[::2]
print(alternate_rows)
```

```python
#task 10 
sales_department = df[df['Department'] == 'Sales']
# Calculate the 90th percentile salary 
top_10_percent_salary = sales_department['Salary'].quantile(0.9)
# Select rows where Salary is in the top 10%
top_10_sales = sales_department[sales_department['Salary'] >= top_10_percent_salary]
print(top_10_sales)
```

```python
#task 11 
df['Status'] = df['Status'].replace('Inactive', 'Retired')
print(df)
```

```python
#task 12 
df['Bonus'] = df['Salary'] * 0.1
print(df)
```

```python
#task 13 
df.loc[(df['Department'] == 'IT') & (df['Performance Score'].isna()), 'Performance Score'] = 5
print(df)
```

```python
#task 14 
df.loc[df['Experience'] < 2, 'Experience'] = 2
print(df)
```

```python
#task 15 
df['Normalized Salary'] = (df['Salary'] - df['Salary'].min()) / (df['Salary'].max() - df['Salary'].min())
print(df)
```

```python
#task 16 
df.loc[df['Performance Score'] < 3, 'Location'] = 'Remote'
print(df)
```

```python
#task 17 
import pandas as pd
import numpy as np
conditions = [(df['Experience'] < 3),               # Junior
    (df['Experience'] >= 3) & (df['Experience'] <= 5),  # Mid
    (df['Experience'] > 5)              ]

experience_levels = ['Junior', 'Mid', 'Senior']
df['Experience Level'] = np.select(conditions, experience_levels, default='Unknown')
print(df)
```

```python
#task 18 
new_row = {'ID': 18,
    'Name': 'areeba ch ',
    'Age': 21,
    'Gender': 'Female',
    'Department': 'Sales',
    'Salary': 75000,
    'Joining Date': '2023-05-01',
    'Performance Score': 4,
    'Experience': 1,
    'Status': 'Active',
    'Location': 'Lahore',
    'Session': 'Morning'}

new_row_df = pd.DataFrame([new_row])

# using pd.concat() as append fucntion is not available here
df = pd.concat([df, new_row_df], ignore_index=True)
print(df)
```

```python
#task 19 
df = df.drop(columns=['Session'])
print(df)
```

```python
#task 20 
df = df.drop(df[df['Age'] > 60].index)
print(df)
```

```python
#task 21 
import pandas as pd
import numpy as np
conditions = [
    (df['Age'] < 30),  # Young
    (df['Age'] >= 30) & (df['Age'] < 50),  # Middle-aged
    (df['Age'] >= 50)  # Senior
]

age_groups = ['Young', 'Middle-aged', 'Senior']
df['Age Group'] = np.select(conditions, age_groups, default='Unknown')
print(df)
```

```python
#task 22 
missing_values = df.isna().sum()
columns_with_missing_values = missing_values[missing_values > 0]
print(columns_with_missing_values)
```

```python
#task 23 
average_score = df['Performance Score'].mean()
df['Performance Score'] = df['Performance Score'].fillna(average_score)
print(df)
```

```python
#task 24 
df = df.dropna()
print(df)
```

```python
#task 25 
median_per_department = df.groupby('Department')['Performance Score'].transform('median')
df['Performance Score'] = df['Performance Score'].fillna(median_per_department)
print(df)
```

```python
#task 26 
average_age_it = df[df['Department'] == 'IT']['Age'].mean()
it_above_avg_age = df[(df['Department'] == 'IT') & (df['Age'] > average_age_it)]
print(it_above_avg_age)
```

```python
#task 27 
duplicates = df[df.duplicated(subset=['Name', 'Joining Date'])]
df = df.drop_duplicates(subset=['Name', 'Joining Date'])
print(duplicates)
```

```python
#task 29
df_sorted_salary = df.sort_values(by='Salary', ascending=False)
print("Task 29:", df_sorted_salary)
```

```python
#task 30 
df_sorted_dept_date = df.sort_values(by=['Department', 'Joining Date'])
print("Task 30:", df_sorted_dept_date)

```

```python
#task 31 
df['Salary Rank'] = df.groupby('Department')['Salary'].rank(ascending=False)
print("Task 31:", df)
```

```python
#task 32 
df_sorted_exp_perf = df.sort_values(by=['Experience', 'Performance Score'], ascending=[False, True])
print("Task 32:", df_sorted_exp_perf)
```

```python
#task 33 
hr_above_5000 = df[(df['Department'] == 'HR') & (df['Salary'] > 5000)]
print("Task 33:", hr_above_5000)
```

```python
#task 34 
df = df[df['Location'] != 'Chicago']
print("Task 34: Updated DataFrame")
```

```python
#task 35 
df = df[df['Performance Score'] >= 2]
print("Task 35: Updated DataFrame")
```

```python
#task 36
avg_salary_per_dept = df.groupby('Department')['Salary'].mean()
print("Task 36:", avg_salary_per_dept)
```

```python
#task 37 
total_salary_status = df.groupby('Status')['Salary'].sum()
print("Task 37:", total_salary_status)
```

```python
#task 38 
max_perf_location = df.groupby('Location')['Performance Score'].max()
print("Task 38:", max_perf_location)
```

```python
#task 39 
df['Joining Year'] = pd.to_datetime(df['Joining Date']).dt.year
total_salary_by_year = df.groupby('Joining Year')['Salary'].sum()
print("Task 39:", total_salary_by_year)
```

```python
#task 40 
avg_perf_dept = df.groupby('Department')['Performance Score'].mean()
highest_perf_dept = avg_perf_dept.idxmax()
print("Task 40:", highest_perf_dept)
```

```python
#task 42
grouped_status_salary = df.groupby('Status')['Salary'].agg(['sum', 'mean', 'std'])
print("Task 42:", grouped_status_salary)
```

```python
#task 43
df['Monthly Salary'] = df['Salary'] / 12
print("Task 43:", df[['Salary', 'Monthly Salary']])
```

```python
#task 44
df_upper = df.applymap(lambda x: x.upper() if isinstance(x, str) else x)
print("Task 44:", df_upper)

```

```python
#task 45 
department_budget = pd.DataFrame({
    'Department': ['IT', 'HR', 'Sales'],
    'Budget': [50000, 30000, 70000]
})
df_inner_join = pd.merge(df, department_budget, on='Department', how='inner')
print("Task 45:", df_inner_join)
```

```python
#task 46
import pandas as pd
recent_projects = pd.DataFrame({
    'ID': df['ID'],
    'Recent Projects': ['ProjA', 'ProjB', 'ProjC', 'ProjD'] * (len(df) // 4)  # Repeat the list
})
recent_projects['Recent Projects'] = recent_projects['Recent Projects'].head(len(df))
df_merged_projects = pd.merge(df, recent_projects, on='ID', how='inner')

print("Task 46:", df_merged_projects)
```

```python
#task 47 
department_leads = pd.DataFrame({
    'Department': ['IT', 'HR', 'Sales'],
    'Team Lead': ['Alice', 'Bob', 'Charlie']
})
df_left_join = pd.merge(df, department_leads, on='Department', how='left')
print("Task 47:", df_left_join)
```

```python
#task 48 
import pandas as pd
project_count = pd.DataFrame({
    'ID': df['ID'],
    'Project Count': [2, 3, 1, 4] * (len(df) // 4)  # Repeat the list enough times
})
project_count['Project Count'] = project_count['Project Count'].head(len(df))
df_project_merge = pd.merge(df, project_count, on='ID')
df_project_merge['Avg Salary Per Project'] = df_project_merge['Salary'] / df_project_merge['Project Count']

print("Task 48:", df_project_merge)

```

```python
#task 49 
df['Dept Avg Salary'] = df.groupby('Department')['Salary'].transform('mean')
df['Salary % Increase'] = ((df['Salary'] - df['Dept Avg Salary']) / df['Dept Avg Salary']) * 100
print("Task 49:", df[['Department', 'Salary', 'Salary % Increase']])

```
